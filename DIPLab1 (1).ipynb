```python
import cv2
import matplotlib.pyplot as plt
img= cv2.imread('/Image02.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
print("img.shape: ", img.shape)
plt.imshow(img)
plt.title("Acquired image")
plt.axis('off')
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.figure(figsize=(5,5))
plt.imshow(gray, cmap='gray')
plt.title("Image Divided into Grid \n(pixel Representation)")
step = 20

for i in range(0, gray.shape[0], step):
  plt.axhline(i, color='orange', linewidth=0.6)

for j in range(0, gray.shape[1], step):
  plt.axvline(j, color='white', linewidth=0.6)

plt.axis('off')
plt.show()
```

```python
print("Pixel intensity values (sample): \n")
print(gray[0:10, 0:10])
```

```python
sampled =cv2.resize(gray, (40,40))

plt.subplot(1,2,1)
plt.imshow(gray, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(sampled, cmap='gray')
plt.title("Sampled Image \n(Reduced Resolution)")
plt.axis('off')

plt.show()
```

```python
import numpy as np
levels=25

quantized= np.floor(gray / (256/levels)) * (256/levels)
quantized = quantized.astype(np.uint8)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(gray, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(quantized, cmap='gray')
plt.title("Quantized Image \n(16 Gray Colors)")
plt.axis('off')

plt.show()

```
